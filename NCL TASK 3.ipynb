```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import lightgbm as lgb
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load data
def load_data(stage_folder, num_rows=None):
    features = []
    for root, dirs, files in os.walk(stage_folder):
        for file in files:
            if file.startswith("f") and file.endswith(".csv"):
                df = pd.read_csv(os.path.join(root, file), header=None)
                if num_rows:
                    df = df.iloc[:num_rows]
                features.append(df.values)
    return np.concatenate(features, axis=1)

# Function to load training data for stages 0 and 1
def load_training_data(stage0_path, stage1_path):
    X_stage0 = load_data(stage0_path)
    X_stage1 = load_data(stage1_path)

    X_train_combined = np.concatenate([X_stage0, X_stage1], axis=0)
    y_train_combined = np.concatenate([np.zeros(X_stage0.shape[0]), np.ones(X_stage1.shape[0])])

    return X_train_combined, y_train_combined

# Function to load testing data for stages 2, 3, and 4
def load_testing_data(stage2_path, stage3_path, stage4_path):
    X_stage2 = load_data(stage2_path)
    X_stage3 = load_data(stage3_path)
    X_stage4 = load_data(stage4_path)

    X_test_combined = np.concatenate([X_stage2, X_stage3, X_stage4], axis=0)
    y_test_combined = np.concatenate([np.full(X_stage2.shape[0], 0),
                                      np.full(X_stage3.shape[0], 1),
                                      np.full(X_stage4.shape[0], 1)])

    return X_test_combined, y_test_combined, len(X_stage2), len(X_stage3), len(X_stage4)

# Train and evaluate model
def train_and_evaluate_model(stage0_path, stage1_path, stage2_path, stage3_path, stage4_path):
    print(f"Training and evaluating model")

    # Load training data
    X_train_combined, y_train = load_training_data(stage0_path, stage1_path)

    # Load testing data
    X_test_combined, y_test_combined, len_stage2, len_stage3, len_stage4 = load_testing_data(stage2_path, stage3_path, stage4_path)

    # Scale the data
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_combined)
    X_test_scaled = scaler.transform(X_test_combined)

    # Convert scaled data to non-negative counts for MultinomialNB
    X_train_counts = np.abs(X_train_scaled)
    X_test_counts = np.abs(X_test_scaled)

    # List of classifiers
    classifiers = {
        'Random Forest': RandomForestClassifier(),
        'Multinomial NB': MultinomialNB(),
        'LightGBM': lgb.LGBMClassifier(),
        'Linear SVM': SVC(kernel='linear', probability=True),
        'Non-Linear SVM': SVC(kernel='rbf', probability=True)
    }

    # Store results for all classifiers
    results = []

    # Train and evaluate each classifier
    for name, clf in classifiers.items():
        print(f"Training {name}")
        if name == 'Multinomial NB':
            clf.fit(X_train_counts, y_train)
            predicted_labels = clf.predict(X_test_counts)
        else:
            clf.fit(X_train_scaled, y_train)
            predicted_labels = clf.predict(X_test_scaled)

        # Classification metrics
        accuracy = accuracy_score(y_test_combined, predicted_labels)
        precision = precision_score(y_test_combined, predicted_labels, average='weighted', zero_division=1)
        recall = recall_score(y_test_combined, predicted_labels, average='weighted', zero_division=1)
        f1 = f1_score(y_test_combined, predicted_labels, average='weighted', zero_division=1)
        results.append((name, accuracy, precision, recall, f1))

        # Confusion Matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(confusion_matrix(y_test_combined, predicted_labels), annot=True, fmt='d', cmap='Blues',
                    xticklabels=["Stage 0", "Stage 1"], yticklabels=["Stage 0", "Stage 1"])
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.title(f'Confusion Matrix for {name}')
        plt.show()

        # Classification report
        print(f"Classification Report for {name}:")
        print(classification_report(y_test_combined, predicted_labels, target_names=["Stage 0", "Stage 1"]))

        # Count predictions for each stage
        predicted_stage2 = predicted_labels[:len_stage2]
        predicted_stage3 = predicted_labels[len_stage2:len_stage2 + len_stage3]
        predicted_stage4 = predicted_labels[len_stage2 + len_stage3:]

        print(f"{name} Predictions for Stage 2: {np.bincount(predicted_stage2.astype(int))}")
        print(f"{name} Predictions for Stage 3: {np.bincount(predicted_stage3.astype(int))}")
        print(f"{name} Predictions for Stage 4: {np.bincount(predicted_stage4.astype(int))}")

    # Results table
    results_df = pd.DataFrame(results, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
    print("Results Summary:")
    print(results_df)

# Define paths to data for EO and EC conditions
paths = {
    "EO": {
        "stage0": "C:/NCL TASK 2/Stage 0/EO",
        "stage1": "C:/NCL TASK 2/Stage 1/EO",
        "stage2": "C:/NCL TASK 2/Stage 2/EO",
        "stage3": "C:/NCL TASK 2/Stage 3/EO",
        "stage4": "C:/NCL TASK 2/Stage 4/EO"
    },
    "EC": {
        "stage0": "C:/NCL TASK 2/Stage 0/EC",
        "stage1": "C:/NCL TASK 2/Stage 1/EC",
        "stage2": "C:/NCL TASK 2/Stage 2/EC",
        "stage3": "C:/NCL TASK 2/Stage 3/EC",
        "stage4": "C:/NCL TASK 2/Stage 4/EC"
    }
}

# Train and evaluate models for EO and EC conditions
for condition, path_dict in paths.items():
    print(f"Condition: {condition}")
    train_and_evaluate_model(path_dict["stage0"], path_dict["stage1"], path_dict["stage2"], path_dict["stage3"], path_dict["stage4"])
```
